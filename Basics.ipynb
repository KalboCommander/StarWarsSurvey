```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

star_wars = pd.read_csv(r'C:\Users\Sam\Python\GuidedProject-StarWarsSurvey\star_wars.csv',encoding = "ISO-8859-1")

star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[pd.notnull(star_wars['RespondentID'])]
star_wars
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

print(star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False))
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```

```python
episode_mapper = {"Star Wars: Episode I  The Phantom Menace":True,
                  "Star Wars: Episode II  Attack of the Clones":True,
                  "Star Wars: Episode III  Revenge of the Sith":True,
                  "Star Wars: Episode IV  A New Hope":True,
                  "Star Wars: Episode V The Empire Strikes Back":True,
                  "Star Wars: Episode VI Return of the Jedi":True,
                  np.NaN: False}

for n in range(3,9):
    star_wars[star_wars.columns[n]] = star_wars[star_wars.columns[n]].map(episode_mapper)
```

```python
sw_columns_rename = {star_wars.columns[3]:"seen_1",
                     star_wars.columns[4]:"seen_2",
                     star_wars.columns[5]:"seen_3",
                     star_wars.columns[6]:"seen_4",
                     star_wars.columns[7]:"seen_5",
                     star_wars.columns[8]:"seen_6",
                    }
star_wars = star_wars.rename(columns = sw_columns_rename)
star_wars.columns
```

```python
star_wars[star_wars.columns[3:9]]
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
sw_ranking_rename = {star_wars.columns[9]:"ranking_1",
                     star_wars.columns[10]:"ranking_2",
                     star_wars.columns[11]:"ranking_3",
                     star_wars.columns[12]:"ranking_4",
                     star_wars.columns[13]:"ranking_5",
                     star_wars.columns[14]:"ranking_6",
                    }
star_wars = star_wars.rename(columns = sw_ranking_rename)
star_wars[star_wars.columns[9:15]]
```

```python
rankings = {}

for n in range(1,7):
    rankings[n] = star_wars['ranking_{}'.format(n)].mean()
    print('ranking_{}:'.format(n),star_wars['ranking_{}'.format(n)].mean())
```

```python
%matplotlib inline
plt.bar(*zip(*rankings.items()))
plt.show()
```

So far we have cleaned up the Star Wars survey to sort through fan preferences amongst all the Star Wars franchise movies.  Since some of the survey results didn't translate well to a dataframe, we needed to rename and recreate some of the data to fit a more uniform structure for exploration and analysis.  Based on our findings, Empire Strikes Back has been the most popular fan favorite among the Star Wars movies.

```python
fans = star_wars[star_wars[star_wars.columns[2]]==True]
non_fans = star_wars[star_wars[star_wars.columns[2]]==False]
```

```python
fan_rankings = {}

for n in range(1,7):
    fan_rankings[n] = fans['ranking_{}'.format(n)].mean()
    print('fan_ranking_{}:'.format(n),fans['ranking_{}'.format(n)].mean())
```

```python
fan_viewership = {}

for n in range(1,7):
    fan_viewership[n] = fans['seen_{}'.format(n)].sum()
    print('fan_views_{}:'.format(n),fans['seen_{}'.format(n)].sum())
```

```python
non_fan_rankings = {}

for n in range(1,7):
    non_fan_rankings[n] = non_fans['ranking_{}'.format(n)].mean()
    print('non_fan_ranking_{}:'.format(n),non_fans['ranking_{}'.format(n)].mean())
```

```python
non_fan_viewership = {}

for n in range(1,7):
    non_fan_viewership[n] = non_fans['seen_{}'.format(n)].sum()
    print('non_fan_views_{}:'.format(n),non_fans['seen_{}'.format(n)].sum())
```

Viewership between fans and non-fans of the star wars franchise run more or less in parallel, however non-fans ranked Episode 1 much higher than fans of the overall franchise did.
